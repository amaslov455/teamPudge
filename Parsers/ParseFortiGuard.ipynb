```
def title_parser(url):
    session = requests.Session()
    base_url = 'http://fortiguard.com/webfilter?q=' + url
    headers = {
    'Accept' : '*/*',
    'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.130 Safari/537.36'
    }
    time.sleep(2)
    try:
      request = session.get(base_url,headers = headers,verify=False, timeout = 123,stream = False)
      if request.status_code == 200:
        print(200)
        soup = bs(request.content,'lxml')
        category = soup.find('h4',attrs={'class':'info_title'}).text
        category = category.split(':')[-1].strip()
        return category
      else:
        print(request.status_code)
        return 'none'
    except Exception as err:
      print(err)
      return 'none'
```

```
if __name__ == '__main__':

    urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)
    count = 0
    urls = df['url_api'].to_list()

    ultrastarttime = time.time()
    dflen = len(df)
    for i in range((dflen//100)+1):
        if i>=0:
            start_time = time.time()
            print(i)
            df_url = df[100*i:100*(i+1)]
            category = map(title_parser, urls[100*i:100*(i+1)])

            df_url['category'] = list(category)
            try:
                if i == 0:
                    with open('api_category.csv', 'w', encoding='utf-8-sig',newline='') as f:
                        df_url.to_csv(f, header=True)
                else:
                    print('Start saving...')
                    with open('api_category.csv', 'a', encoding='utf-8-sig',newline='') as f:
                        df_url.to_csv(f, header=False)
            except:
                pass

            print(f"{(time.time() - start_time)} секунд")
```
